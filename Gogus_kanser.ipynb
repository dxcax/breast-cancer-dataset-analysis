## Model Eğitme  

```python
prediction_vars = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean','concave points_mean','radius_se',
                   'area_se','radius_worst','perimeter_worst','compactness_worst']
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.15, random_state=1)
```

```python
train_x = train[prediction_vars]   # Eğitim için özellikler (input)
train_y = train['diagnosis']       # Eğitim için hedef değişken (label)
test_x = test[prediction_vars]     # Test için özellikler (input)
test_y = test['diagnosis']         # Test için gerçek sonuçlar (label)
```

```python
from sklearn.neural_network import MLPClassifier     # Çok katmanlı yapay sinir ağı
from sklearn.ensemble import RandomForestClassifier  # Rastgele orman (çoklu karar ağacı)
from sklearn.neighbors import KNeighborsClassifier    # K-en yakın komşu algoritması
from sklearn.svm import SVC                           # Destek vektör makineleri (SVM)
```

```python
model = RandomForestClassifier()     # Random Forest sınıflandırıcı modelini oluşturuyoruz
```

```python
model.fit(train_x, train_y)          # Modeli eğitim verisi (özellikler ve hedef) ile eğitiyoruz
```

```python
predictions = model.predict(test_x)  # Test verisi ile tahmin yap
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix     # Gerçek ve tahmin sonuçlarla karışıklık matrisini oluştur
confusion_matrix(test_y, predictions)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision = precision_score(test_y, predictions) #precision=kesinlik
print("Kesinlik değeri %.2f" % precision)

recall = recall_score (test_y, predictions)  #recall = duyarlılık
print("Duyarlılık değeri %.2f" % recall)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(test_y, predictions)            # Doğruluk (accuracy) hesapla
print("The accuracy is %.2f" % accuracy)                  # Doğruluk değerini yazdır
```
